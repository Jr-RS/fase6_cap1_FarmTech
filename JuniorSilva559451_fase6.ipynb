## AnÃ¡lise dos resultados apresentados

```python
# Caminhos para os arquivos results.csv de cada modelo
caminhos_results = [
    '/content/yolov5/runs/train/treino_30epocas/results.csv',
    '/content/yolov5/runs/train/treino_45epocas/results.csv',
    '/content/yolov5/runs/train/treino_60epocas/results.csv'
]

nomes_modelos = ['30 Ã©pocas', '45 Ã©pocas', '60 Ã©pocas']

# Listas para armazenar mÃ©tricas finais
precisao_final = []
revocacao_final = []
map50_final = []

# Loop para cada modelo
for i, caminho in enumerate(caminhos_results):
  print(f"Analisando Resultados do Modelo {nomes_modelos[i]}")
  # Ler o CSV
  df_results = pd.read_csv(caminho)
  df_results.columns = df_results.columns.str.strip()  # Corrigir espaÃ§os

  # GrÃ¡fico 1 - EvoluÃ§Ã£o dos Losses
  plt.figure(figsize=(10, 6))
  plt.plot(df_results['epoch'], df_results['train/box_loss'], label='Train Box Loss')
  plt.plot(df_results['epoch'], df_results['val/box_loss'], label='Val Box Loss')
  plt.plot(df_results['epoch'], df_results['train/obj_loss'], label='Train Obj Loss')
  plt.plot(df_results['epoch'], df_results['val/obj_loss'], label='Val Obj Loss')
  plt.plot(df_results['epoch'], df_results['train/cls_loss'], label='Train Cls Loss')
  plt.plot(df_results['epoch'], df_results['val/cls_loss'], label='Val Cls Loss')
  plt.xlabel('Ã‰poca')
  plt.ylabel('Loss')
  plt.title(f'EvoluÃ§Ã£o dos Losses ({nomes_modelos[i]})')
  plt.legend()
  plt.grid(True)
  plt.show()

  # GrÃ¡fico 2 - EvoluÃ§Ã£o dos mAPs
  plt.figure(figsize=(10, 6))
  plt.plot(df_results['epoch'], df_results['metrics/mAP_0.5'], label='mAP@0.5')
  plt.plot(df_results['epoch'], df_results['metrics/mAP_0.5:0.95'], label='mAP@0.5:0.95')
  plt.xlabel('Ã‰poca')
  plt.ylabel('mAP')
  plt.title(f'EvoluÃ§Ã£o do mAP ({nomes_modelos[i]})')
  plt.legend()
  plt.grid(True)
  plt.show()

  # GrÃ¡fico 3 - EvoluÃ§Ã£o de PrecisÃ£o e Recall
  plt.figure(figsize=(10, 6))
  plt.plot(df_results['epoch'], df_results['metrics/precision'], label='PrecisÃ£o')
  plt.plot(df_results['epoch'], df_results['metrics/recall'], label='Recall')
  plt.xlabel('Ã‰poca')
  plt.ylabel('Valor')
  plt.title(f'EvoluÃ§Ã£o da PrecisÃ£o e Recall ({nomes_modelos[i]})')
  plt.legend()
  plt.grid(True)
  plt.show()

  # Coletar a Ãºltima linha de mÃ©tricas (Ãºltima Ã©poca)
  precisao_final.append(df_results['metrics/precision'].iloc[-1])
  revocacao_final.append(df_results['metrics/recall'].iloc[-1])
  map50_final.append(df_results['metrics/mAP_0.5'].iloc[-1])

  print('')

# GrÃ¡fico final comparando os modelos
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.bar(nomes_modelos, precisao_final, color='blue')
plt.title('PrecisÃ£o Final')
plt.ylim(0, 1)
plt.grid(axis='y')

plt.subplot(1, 3, 2)
plt.bar(nomes_modelos, revocacao_final, color='green')
plt.title('RevocaÃ§Ã£o (Recall) Final')
plt.ylim(0, 1)
plt.grid(axis='y')

plt.subplot(1, 3, 3)
plt.bar(nomes_modelos, map50_final, color='orange')
plt.title('mAP@0.5 Final')
plt.ylim(0, 1)
plt.grid(axis='y')

plt.suptitle('ComparaÃ§Ã£o dos Modelos Treinados')
plt.tight_layout()
plt.show()
```

## Limpar ambiente

```python
def delete_folder_contents(folder_path):
  """Deletes all contents of a folder."""
  if os.path.exists(folder_path):
    for filename in os.listdir(folder_path):
      file_path = os.path.join(folder_path, filename)
      try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
          os.unlink(file_path)
        elif os.path.isdir(file_path):
          shutil.rmtree(file_path)
      except Exception as e:
        print(f'Failed to delete {file_path}. Reason: {e}')
  else:
    print(f"Folder '{folder_path}' not found.")

# Delete contents of the specified folders
#delete_folder_contents('/content/yolov5/runs/detect')
#delete_folder_contents('/content/yolov5/runs/train')
```
